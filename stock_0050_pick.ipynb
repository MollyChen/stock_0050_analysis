```python
import pandas as pd
import sqlite3
import os
import numpy as np
import stock_0050_func
```

```python
conn = sqlite3.connect(os.path.join('data','data.db'))

df = pd.read_sql('select 日期, 收盤價, 開盤價, 最高價, 最低價, 成交股數 from day_price',conn, index_col=['日期'], parse_dates=['日期'])
df.rename(columns={'日期':'date', '收盤價':'close', '開盤價':'open','最高價':'high','最低價':'low', '成交股數':'volume'},inplace=True)
df = df.apply(lambda s: pd.to_numeric(s, errors='coerce'))
df.head(10)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from talib import abstract

plt.rcParams['figure.figsize'] = (18,10)

abstract.SMA(df).plot(label='SMA')
df['close'].plot(label='close')
plt.legend()
```
